```python
print(jobs_results)
```

```python
import csv

  
with open('output.csv', 'w', newline='') as csvfile:
	csv_writer = csv.writer(csvfile)
	
	# Write the headers
	csv_writer.writerow(["title", "company_name", "Snippet"])
	
	# Write the data
	for result in jobs_results:
		csv_writer.writerow([result["title"], result["via"],result["share_link"],result["description"]])

  
print('Done writing to CSV file.')
```
